**1.** Use el m茅todo de bisecci贸n para encontrar soluciones precisas dentro de  $ 10^{-2}$ para $ x^{3} -7x^{2} + 14x-6=0$  en cada intervalo.


```python
def f(x):                               # Se define la funci贸n
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(a, b, tol=1e-2, max_iter=50): # M茅todo de bisecci贸n
    print(f"\nIntervalo inicial: [{a}, {b}]")
    
    if f(a) * f(b) >= 0:    # Verifica que f(a) y f(b) tengan signos opuestos
        print("No aplica, f(a) y f(b) deben tener signos opuestos.")
        return None

    iteracion = 0   # Inicializa contador de iteraciones.
    while abs(b - a) > tol and iteracion < max_iter:    
        p = (a + b) / 2.0                               
        fa, fb, fc = f(a), f(b), f(p)

        print(f"Iteraci贸n {iteracion + 1}: a = {a:.6f}, b = {b:.6f}, p = {p:.6f}, "
              f"f(a) = {fa:.6f}, f(b) = {fb:.6f}, f(p) = {fc:.6f}")

        if fc == 0: # Si f(p) = 0, se encontr贸 la ra铆z exacta y se retorna.
            print(f"Ra铆z exacta encontrada en p = {p:.6f}")
            return p  # Se encontr贸 la ra铆z exacta

        if fa * fc < 0:
            b = p
        else:
            a = p

        iteracion += 1

    raiz_aprox = (a + b) / 2.0
    print(f"Ra铆z aproximada: {raiz_aprox:.6f}")
    print(f"Iteraciones totales: {iteracion}\n")
    return raiz_aprox
```

**Suponga que  = 10 ,  = 1  y  = 12.4 $cm^{3} $. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 .**

$ V = L[0.5\pi r^{2} - r^{2}arcsen(h/r) -h(r^2 - h^2)^{\frac{1}{2}}] $
- Reemplazando valores:

$ 12.4 = 10[0.5\pi r^{2} - r^{2}arcsen(h/r) -h(r^2 - h^2)^{\frac{1}{2}}] $

$f(h) = 12.4 -10[0.5\pi r^{2} - r^{2}arcsen(h/r) -h(r^2 - h^2)^{\frac{1}{2}}] $

Intervalo [a,b]=[0,1]

$f(0) = 12.4 - 10[0.5\pi - sen^{-1}(0)-0] = -3.308$
- Para f(1)

Luego,

$f(a)f(b) = f(0)f(1) = (-3.308)(12.4) = -41.0192 < 0$


```python
import numpy as np

L = 10        
r = 1         
Vol = 12.4  
tolerance = 0.01  

def V(h):
    return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h / r) - h * np.sqrt(r**2 - h**2))

def f(h):
    return V(h) - Vol

def biseccion_volumen(a, b, tol=1e-2, max_iter=100):
    if f(a) * f(b) >= 0:
        print("No aplica. f(a) y f(b) deben tener signos opuestos.")
        print(f"f(a) = {f(a):.6f}, f(b) = {f(b):.6f}")
        return None

    iteracion = 1
    while abs(b - a) >= tol and iteracion <= max_iter:
        p = (a + b) / 2
        fp = f(p)

        print(f"Iteraci贸n {iteracion}: a = {a:.5f}, b = {b:.5f}, p = {p:.5f}, f(p) = {fp:.6f}")

        # Condici贸n de paro
        if abs(fp) < 1e-6:
            break
        elif f(a) * fp < 0:
            b = p
        else:
            a = p

        iteracion += 1

    return (a + b) / 2

h_sol = biseccion_volumen(0.0001, r - 0.0001, tolerance)
if h_sol is not None:
    print(f"\nh =: {h_sol:.3f} cm")
    print(f"Volumen calculado = {V(h_sol):.4f} cm鲁, error = {abs(V(h_sol) - Vol):.4f} cm鲁")
```
