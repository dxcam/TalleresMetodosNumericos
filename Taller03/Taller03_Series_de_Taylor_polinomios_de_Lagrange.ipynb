## **SERIES DE TAYLOR**

$$P_n(x) = f(x_0) + f'(x_0)(x-x_0) + \frac{f''(x_0)}{2!}(x-x_0)^2 + ... + \frac{f^{n}(x_0)}{n!}(x-x_0)^n = \sum_{k=0}^{n}\frac{f^{(k)}(x_0)}{k!}(x-x_0)^k$$

**1.** $cos(x), x_0 = 0$


- $f(x) = cos(x)$
- $f'(x) = -sen(x)$
- $f''(x) = -cos(x)$
- $f'''(x) = sen(x)$
- $f^4(x) = cos(x)$

$P_4(x) = cos(0) - sen(0)(x-0) + \frac{-cos(0)}{2!}(x-0)^2 + \frac{sen(0)}{3!}(x-0)^3 + \frac{cos(0)}{4!}(x-0)^4$

$P_4(x) = 1 - 0(x) + \frac{-1}{2}(x)^2 + \frac{0}{6}(x)^3 + \frac{1}{24}(x)^4$

$P_4(x) = \frac{1}{24}x^4 - \frac{1}{2}x^2 + 1$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# FunciÃ³n para calcular la serie de Taylor de cos(x) centrada en x0=0
def serie_taylor(x, n):

    result = 0
    for k in range(n + 1):
        result += ((-1)**k * x**(2*k)) / math.factorial(2*k)
    return result

x = np.linspace(-2*np.pi, 2*np.pi, 400)

plt.figure(figsize=(12, 7))
plt.plot(x, np.cos(x), label='cos(x) real', color='black', linewidth=2.5)   # Graficar cos(x) 

# Graficar aproximaciones de Taylor de distintos Ã³rdenes
orders = [0, 1, 2, 4, 6]
colors = plt.cm.tab10(np.linspace(0, 1, len(orders)))

for idx, n in enumerate(orders):
    y_taylor = np.array([serie_taylor(val, n) for val in x])
    plt.plot(x, y_taylor, label=f'Taylor orden {n}', linewidth=1.5, alpha=0.8, color=colors[idx])

plt.title('Aproximaciones de Taylor para cos(x) con xâ=0', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.ylim(-2, 2)
plt.tight_layout()
plt.show()
```

**2.** $\frac{1}{1-x}, x_0= 0.5 $

- $f(x) = \frac{1}{1-x}$
- $f'(x) = \frac{1}{(1-x)^2}$
- $f''(x) = \frac{2}{(1-x)^3}$
- $f'''(x) = \frac{6}{(1-x)^4}$
- $f^4(x) = \frac{24}{(1-x)^5}$

$P_4(x) = \frac{1}{1-0.5} + \frac{1}{(1-0.5)^2}(x-0.5) + \frac{\frac{2}{(1-0.5)^3}}{2!}(x-0.5)^2 + \frac{\frac{6}{(1-0.5)^4}}{3!}(x-0.5)^3 + \frac{\frac{24}{(1-x)^5}}{4!}(x-0.5)^4$

$P_4(x) = 2 + 4(x-0.5) + 8(x-0.5)^2 + 16(x-0.5)^3 + 32(x-0.5)^4$



```python
import numpy as np
import matplotlib.pyplot as plt

def serie_taylor(x, x0, n):
    # f(x)=1/(1-x)=sum_{k=0}^n (x-x0)^k / (1-x0)^{k+1}
    x = np.asarray(x)
    resultado = np.zeros_like(x, dtype=float)
    denominador = 1 - x0
    for k in range(0, n+1):
        resultado += (x - x0)**k / (denominador**(k+1))
    return resultado

x0 = 0.5
x = np.linspace(0.0, 0.999, 400)
y_true = 1.0 / (1.0 - x)

plt.figure(figsize=(12, 7))
plt.plot(x, y_true, label='f(x)=1/(1-x) real', color='black', linewidth=2)

orders = [0, 1, 2, 3, 4 ]
colors = plt.cm.viridis(np.linspace(0, 1, len(orders)))
for idx, n in enumerate(orders):
    y_t = serie_taylor(x, x0, n)
    plt.plot(x, y_t, label=f'Taylor orden {n}', color=colors[idx], linewidth=1.5)

plt.title('Series de Taylor para f(x)=1/(1-x) con x0=0.5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(0, 20)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

**3.** $In(x), x_0 = 1$

- $f(x) = In(x)$
- $f'(x) = \frac{1}{x}$
- $f''(x) = -\frac{1}{x^2}$
- $f'''(x) = \frac{2}{x^3}$
- $f^4(x) = -\frac{6}{x^4}$

$P_4(x) = In(1) + \frac{1}{1}(x-1) - \frac{\frac{1}{1^2}}{2!}(x-1)^2 + \frac{\frac{2}{1^3}}{3!}(x-1)^3 - \frac{\frac{6}{1^4}}{4!}(x-1)^4 $

$P_4(x) = 0 + (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4 $

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# FunciÃ³n para calcular la serie de Taylor de ln(x) centrada en x0=1
def serie_taylor(x, n):
    if x <= 0:
        return float('nan')  # ln(x) no estÃ¡ definido para x <= 0
    z = x - 1
    result = 0
    for k in range(1, n + 1):
        result += ((-1)**(k + 1)) * (z**k) / k
    return result

x = np.linspace(0.1, 2.5, 400)

plt.plot(x, np.log(x), label='ln(x)', color='black', linewidth=2)   # Graficar ln(x) real

orders = [1, 2, 3, 4, 5, 6]
colors = plt.cm.viridis(np.linspace(0, 1, len(orders)))

for n in orders:
    y_taylor = [serie_taylor(val, n) for val in x]
    plt.plot(x, y_taylor, label=f'Taylor orden {n}', linewidth=1.5)

plt.title('Aproximaciones de Taylor para ln(x) con x=1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

**4.** $In(x), x_0 = 10$

- $f(x) = In(x)$
- $f'(x) = \frac{1}{x}$
- $f''(x) = -\frac{1}{x^2}$
- $f'''(x) = \frac{2}{x^3}$
- $f^4(x) = -\frac{6}{x^4}$

$P_4(x) = In(10) + \frac{1}{10}(x-10) - \frac{\frac{1}{10^2}}{2!}(x-10)^2 + \frac{\frac{2}{10^3}}{3!}(x-10)^3 - \frac{\frac{6}{10^4}}{4!}(x-10)^4 $

$P_4(x) = 2.302585093 + \frac{1}{10}(x-10) - \frac{1}{200} (x-10)^2 + \frac{1}{3000}(x-10)^3 - \frac{1}{40000}(x-10)^4 $

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# FunciÃ³n para calcular la serie de Taylor de ln(x) centrada en x0=10
def serie_taylor(x, x0, n):
    if x <= 0:
        return float('nan')  # ln(x) no estÃ¡ definido para x <= 0
    z = x - x0
    result = math.log(x0)
    for k in range(1, n + 1):
        result += ((-1)**(k + 1)) * (z**k) / (k * x0**k)
    return result

x0 = 10
x = np.linspace(5, 25, 400)
y_true = np.log(x)

plt.figure(figsize=(12, 7))
plt.plot(x, y_true, label='ln(x) real', color='black', linewidth=2.5)

orders = [1, 2, 3, 4, 5, 6]
colors = plt.cm.viridis(np.linspace(0, 1, len(orders)))

for idx, n in enumerate(orders):
    y_taylor = np.array([serie_taylor(val, x0, n) for val in x])
    plt.plot(x, y_taylor, label=f'Taylor orden {n}', color=colors[idx], linewidth=1.5, alpha=0.8)

plt.title('Series de Taylor para f(x)=ln(x) con xâ=10', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## **POLINOMIO DE LAGRANGE**

**Encuentre el polinomio de Lagrange para los siguientes datos y grafique:**

$$ P(x) = f(x_0)L_0(x) + ... + f(x_n)L_n(x) $$

$$ P(x) = y_0L_0(x) + y_1L_1(x) + ... + y_nL_n(x) = \sum_{k=0}^{n} f(x_k)L_k(x) $$

$$L_k(x) = \frac{(x-x_0)(x-x_1)...(x-x_{k-1})(x-x_{k+1})...(x-x_n)}{(x_k-x_0)(x_k-x_1)...(x_k-x_{k-1})(x_k-x_{k+1})...(x_k-x_n)} = \prod _{i=0}^{n} \frac{x-x_i}{x_k-x_i} $$

**1.** $(0,0),(30,0.5),(60,\frac{\sqrt{3}}{2}),(90,1)$

El polinomio es de grado **n = 3**.

$$L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)} = \frac{(x - 30)(x - 60)(x - 90)}{(0 - 30)(0 - 60)(0 - 90)} = \frac{(x - 30)(x - 60)(x - 90)}{(-30)(-60)(-90)} = \frac{(x - 30)(x - 60)(x - 90)}{-162000}$$

$$L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)} = \frac{(x - 0)(x - 60)(x - 90)}{(30 - 0)(30 - 60)(30 - 90)} = \frac{x(x - 60)(x - 90)}{(30)(-30)(-60)} = \frac{x(x - 60)(x - 90)}{54000}$$

$$L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)} = \frac{(x - 0)(x - 30)(x - 90)}{(60 - 0)(60 - 30)(60 - 90)} = \frac{x(x - 30)(x - 90)}{(60)(30)(-30)} = \frac{x(x - 30)(x - 90)}{-54000}$$

$$L_3(x) = \frac{(x - x_0)(x - x_1)(x - x_2)}{(x_3 - x_0)(x_3 - x_1)(x_3 - x_2)} = \frac{(x - 0)(x - 30)(x - 60)}{(90 - 0)(90 - 30)(90 - 60)}= \frac{x(x - 30)(x - 60)}{(90)(60)(30)} = \frac{x(x - 30)(x - 60)}{162000}$$

$$P_3(x) = y_0 L_0(x) + y_1 L_1(x) + y_2 L_2(x) + y_3 L_3(x)$$

$$P_3(x) = (0)L_0(x) + \left(\frac{1}{2}\right) L_1(x) + \left(\frac{\sqrt{3}}{2}\right) L_2(x) + (1) L_3(x)$$
$$P_3(x) = \frac{1}{2} \left[ \frac{x(x - 60)(x - 90)}{54000} \right] + \frac{\sqrt{3}}{2} \left[ \frac{x(x - 30)(x - 90)}{-54000} \right] + \frac{x(x - 30)(x - 60)}{162000}$$
$$P_3(x) = \frac{x(x - 60)(x - 90)}{108000} - \frac{\sqrt{3} x(x - 30)(x - 90)}{108000} + \frac{x(x - 30)(x - 60)}{162000}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import polynomial as P

x_data = np.array([0, 30, 60, 90])
y_data = np.array([0, 0.5, np.sqrt(3)/2, 1])

def lagrange(x, x_data, y_data):    # FunciÃ³n para calcular el polinomio de Lagrange
   
    n = len(x_data)
    resultado = 0
    for i in range(n):
        L_i = 1 # Calcular la base de Lagrange Li(x)
        for j in range(n):
            if i != j:
                L_i *= (x - x_data[j]) / (x_data[i] - x_data[j])
        resultado += y_data[i] * L_i
    return resultado

# Crear puntos para graficar el polinomio
x_continuo = np.linspace(-5, 95, 500)
y_continuo = np.array([lagrange(x, x_data, y_data) for x in x_continuo])

plt.figure(figsize=(12, 7))

# Graficar el polinomio de Lagrange
plt.plot(x_continuo, y_continuo, 'm-', label='Polinomio de Lagrange', linewidth=2.5)

# Graficar los puntos de interpolaciÃ³n
plt.plot(x_data, y_data, 'ro', markersize=10, label='Puntos de datos', zorder=5)

# AÃ±adir anotaciones a los puntos
for i, (xi, yi) in enumerate(zip(x_data, y_data)):
    if i == 2:
        label = f'({xi}, â3/2)'
    else:
        label = f'({xi}, {yi})'
    plt.annotate(label, xy=(xi, yi), xytext=(5, 10), 
                textcoords='offset points', fontsize=10, 
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.title('Polinomio de InterpolaciÃ³n de Lagrange', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.tight_layout()
plt.show()
```

$$L_0(x) = \frac{(x - x_1)(x - x_2)}{(x_0 - x_1)(x_0 - x_2)} = \frac{(x - 2)(x - 3)}{(1 - 2)(1 - 3)} = \frac{(x - 2)(x - 3)}{(-1)(-2)} = \frac{(x^2 - 5x + 6)}{2}$$

$$L_1(x) = \frac{(x - x_0)(x - x_2)}{(x_1 - x_0)(x_1 - x_2)} = \frac{(x - 1)(x - 3)}{(2 - 1)(2 - 3)} = \frac{(x - 1)(x - 3)}{(1)(-1)} = \frac{(x^2 - 4x + 3)}{-1} = -(x^2 - 4x + 3)$$

$$L_2(x) = \frac{(x - x_0)(x - x_1)}{(x_2 - x_0)(x_2 - x_1)} = \frac{(x - 1)(x - 2)}{(3 - 1)(3 - 2)} = \frac{(x - 1)(x - 2)}{(2)(1)} = \frac{(x^2 - 3x + 2)}{2}$$

$$P_2(x) = (1) L_0(x) + (2) L_1(x) + (2) L_2(x)$$
$$P_2(x) = 1 \cdot \left(\frac{x^2 - 5x + 6}{2}\right) + 2 \cdot \left(-(x^2 - 4x + 3)\right) + 2 \cdot \left(\frac{x^2 - 3x + 2}{2}\right)$$
$$P_2(x) = \frac{1}{2} (x^2 - 5x + 6) - 2(x^2 - 4x + 3) + 1(x^2 - 3x + 2)$$

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([1, 2, 3])
y_data = np.array([1, 2, 2])

def lagrange(x, x_data, y_data): # FunciÃ³n para calcular el polinomio de Lagrange

    n = len(x_data)
    resultado = 0
    for i in range(n):
        L_i = 1 # Calcular la base de Lagrange Li(x)
        for j in range(n):
            if i != j:
                L_i *= (x - x_data[j]) / (x_data[i] - x_data[j])
        resultado += y_data[i] * L_i
    return resultado

x_continuo = np.linspace(0.5, 3.5, 500)
y_continuo = np.array([lagrange(x, x_data, y_data) for x in x_continuo])

plt.figure(figsize=(12, 7))

# Graficar el polinomio de Lagrange
plt.plot(x_continuo, y_continuo, 'c-', label='Polinomio de Lagrange Pâ(x)', linewidth=2.5)

# Graficar los puntos de interpolaciÃ³n
plt.plot(x_data, y_data, 'ro', markersize=10, label='Puntos de datos', zorder=5)

# AÃ±adir anotaciones a los puntos
for i, (xi, yi) in enumerate(zip(x_data, y_data)):
    label = f'({xi}, {yi})'
    plt.annotate(label, xy=(xi, yi), xytext=(5, 10), 
                textcoords='offset points', fontsize=10, 
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.title('Polinomio de InterpolaciÃ³n de Lagrange para (1,1), (2,2), (3,2)', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.tight_layout()
plt.show()
```
