**3)** Dir√≠jase al pseudoc√≥digo del spline c√∫bico con frontera natural provisto en clase, en
base a ese pseudoc√≥digo complete la siguiente funci√≥n:



```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol
from sympy import lambdify

def cubic_spline(xs: list[float], ys: list[float]) -> list[sym.Expr]:
    """
    Construye el spline c√∫bico natural que interpola los puntos dados.
    Devuelve una lista de expresiones simb√≥licas S_j(x) v√°lidas en [x_j, x_{j+1}].
    """
    # Ordenar puntos por abscisa (por si no vienen ordenados)
    points = sorted(zip(xs, ys), key=lambda p: p[0])
    xs = [float(x) for x, _ in points]
    ys = [float(y) for _, y in points]

    n = len(xs) - 1
    if n < 1:
        raise ValueError("Se necesitan al menos dos puntos para construir splines.")

    # Calcular h
    h = [xs[i+1] - xs[i] for i in range(n)]

    # Construir vector alpha (lado derecho)
    alpha = np.zeros(n+1, dtype=float)
    for i in range(1, n):
        alpha[i] = 3 * ((ys[i+1] - ys[i]) / h[i] - (ys[i] - ys[i-1]) / h[i-1])

    # Construir y resolver el sistema tridiagonal A * c = alpha
    # A de tama√±o (n+1)x(n+1)
    A = np.zeros((n+1, n+1), dtype=float)
    A[0,0] = 1.0
    A[n,n] = 1.0
    for i in range(1, n):
        A[i, i-1] = h[i-1]
        A[i, i]   = 2 * (h[i-1] + h[i])
        A[i, i+1] = h[i]

    # Ccondiciones naturales: c0 = 0, cn = 0 impl√≠citas por filas 0 y n
    c = np.linalg.solve(A, alpha)  # c es array de longitud n+1

    # Paso 4: calcular b_j y d_j para cada intervalo
    x = Symbol('x')
    splines = []
    for j in range(n):
        a_j = ys[j]
        # F√≥rmulas est√°ndar
        b_j = (ys[j+1] - ys[j]) / h[j] - h[j] * (2*c[j] + c[j+1]) / 3.0
        d_j = (c[j+1] - c[j]) / (3.0 * h[j])
        c_j = c[j] 
        # Construir expresi√≥n simb√≥lica
        S_j = sym.nsimplify(a_j) + sym.nsimplify(b_j) * (x - sym.nsimplify(xs[j])) \
              + sym.nsimplify(c_j)*(x-sym.nsimplify(xs[j]))**2+sym.nsimplify(d_j)*(x-sym.nsimplify(xs[j]))**3
        splines.append(sym.simplify(S_j))

    return splines

# ----------------- EJEMPLO de uso y gr√°fico -----------------
if __name__ == "__main__":
    xs_example = [0, 1, 2]
    ys_example = [1, 5, 3]

    # Calcular splines (natural)
    spls = cubic_spline(xs_example, ys_example)

    # Mostrar las expresiones simb√≥licas
    print("Expresiones de los segmentos del spline c√∫bico natural:")
    for i, S in enumerate(spls):
        print(f"S_{i}(x) = {sym.pretty(S)}\n")

    # Graficar los segmentos evaluando las expresiones simb√≥licas con lambdify
    x_sym = Symbol('x')
    colors = plt.cm.viridis(np.linspace(0, 1, len(spls)))
    plt.figure(figsize=(8, 5))

    for i, S in enumerate(spls):
        f_num = lambdify(x_sym, S, 'numpy')
        x_seg = np.linspace(xs_example[i], xs_example[i+1], 200)
        y_seg = f_num(x_seg)
        plt.plot(x_seg, y_seg, color=colors[i], 
                 label=f'Segmento {i}: [{xs_example[i]}, {xs_example[i+1]}]')
    
    plt.scatter(xs_example, ys_example, color='red', zorder=5, label='Puntos')

    plt.title('Spline c√∫bico natural que interpola los puntos dados')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
```

```python
import sympy as sym
from IPython.display import display, Math
import numpy as np
import matplotlib.pyplot as plt

# Usar la funci√≥n cubic_spline definida anteriormente
xs = [1, 2, 3]
ys = [2, 3, 5]

splines = cubic_spline(xs=xs, ys=ys)

# Mostrar expresiones simb√≥licas
x = sym.Symbol('x')
for i, s in enumerate(splines):
    print(f"\nSegmento {i}: x ‚àà [{xs[i]}, {xs[i+1]}]")
    print(f"S_{i}(x) = {s}")
    display(Math(f"S_{{{i}}}(x) = {sym.latex(s)}"))
    print(f"Forma expandida: {s.expand()}")

# Graficar
from sympy import lambdify

fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.viridis(np.linspace(0, 1, len(splines)))
for i, S in enumerate(splines):
    f_num = lambdify(x, S, 'numpy')
    x_seg = np.linspace(xs[i], xs[i+1], 200)
    y_seg = f_num(x_seg)
    ax.plot(x_seg, y_seg, color=colors[i], linewidth=2.5,
             label=f'S_{i}(x) en [{xs[i]}, {xs[i+1]}]')

# Puntos de interpolaci√≥n
ax.scatter(xs, ys, color='red', s=100, zorder=5, label='Puntos')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('S(x)', fontsize=12)
ax.set_title('Spline C√∫bico Natural: xs = [1, 2, 3], ys = [2, 3, 5]', fontsize=14, 
             fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cubic_spline_clamped_numeric(xs, ys, d0, dn):
    """Calcula coeficientes num√©ricos del spline c√∫bico clamped.
    Retorna arrays (a,b,c,d) tal que en el intervalo i:
      S_i(x) = a[i] + b[i]*(x-x_i) + c[i]*(x-x_i)**2 + d[i]*(x-x_i)**3
    """
    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)
    n = len(xs) - 1
    if n < 1:
        raise ValueError("Se necesitan al menos dos puntos")

    h = xs[1:] - xs[:-1]

    # Construir A y alpha para condiciones clamped
    A = np.zeros((n+1, n+1), dtype=float)
    alpha = np.zeros(n+1, dtype=float)

    A[0,0] = 2*h[0]; A[0,1] = h[0]
    A[n,n-1] = h[n-1]; A[n,n] = 2*h[n-1]

    alpha[0] = 3*((ys[1]-ys[0])/h[0] - d0)
    alpha[n] = 3*(dn - (ys[n]-ys[n-1])/h[n-1])

    for i in range(1,n):
        A[i,i-1] = h[i-1]
        A[i,i]   = 2*(h[i-1] + h[i])
        A[i,i+1] = h[i]
        alpha[i] = 3*((ys[i+1]-ys[i])/h[i] - (ys[i]-ys[i-1])/h[i-1])

    # segundas derivadas en puntos
    m = np.linalg.solve(A, alpha)

    # Coeficientes 
    a = ys[:-1].copy()
    b = np.zeros(n)
    c = np.zeros(n)
    d = np.zeros(n)
    for i in range(n):
        b[i] = (ys[i+1]-ys[i])/h[i] - (h[i]/6)*(2*m[i] + m[i+1])
        c[i] = m[i] / 2.0
        d[i] = (m[i+1] - m[i]) / (6.0*h[i])

    return a, b, c, d


def expresion_polinomica(xi, a, b, c, d):
    """Devuelve coeficientes A,B,C,D de la forma Ax^3 + Bx^2 + Cx + D explicitos num√©ricos."""
    # S(x) = a + b*(x-xi) + c*(x-xi)^2 + d*(x-xi)^3
    A = d
    B = -3*d*xi + c
    C = 3*d*xi**2 - 2*c*xi + b
    D = -d*xi**3 + c*xi**2 - b*xi + a
    return A, B, C, D


def spline_segmento(xs, a, b, c, d, curve_name):
    xs = np.array(xs, dtype=float)
    for i in range(len(a)):
        xi = xs[i]
        xj = xs[i+1]
        ai, bi, ci, di = a[i], b[i], c[i], d[i]
        print(f"\nSegmento {i+1}: intervalo [{xi}, {xj}]")
        print(f"Polinomio: S_{i}(x) = {ai:.6g}+{bi:.6g}*(x-{xi:.6g})+{ci:.6g}*(x-{xi:.6g})**2+{di:.6g}*(x-{xi:.6g})**3")
        A,B,C,D = expresion_polinomica(xi, ai, bi, ci, di)
        print(f"Forma expandida: S_{i}(x) = {A:.6g}*x^3 + {B:.6g}*x^2 + {C:.6g}*x + {D:.6g}")


def plot_numeric_spline(xs, ys, a, b, c, d, title, figsize=(10,5)):
    xs = np.array(xs, dtype=float)
    plt.figure(figsize=figsize)
    plt.plot(xs, ys, 'o', color='red', label='Datos')

    for i in range(len(a)):
        x0 = xs[i]
        x1 = xs[i+1]
        xr = np.linspace(x0, x1, 200)
        dx = xr - x0
        yr = a[i] + b[i]*dx + c[i]*dx**2 + d[i]*dx**3
        plt.plot(xr, yr, label=f'Segmento {i+1} [{x0},{x1}]')

    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# -------------------- DATOS (tabla) --------------------
xs1 = [1, 2, 5, 6, 7, 8, 10, 13, 17]
ys1 = [3.0, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5]
B0_1 = 1.0
B1_1 = -0.67

xs2 = [17, 20, 23, 24, 25, 27, 27.7]
ys2 = [4.5, 7.0, 6.1, 5.6, 5.8, 5.2, 4.1]
B0_2 = 3.0
B1_2 = -4.0

xs3 = [27.7, 28, 29, 30]
ys3 = [4.1, 4.3, 4.1, 3.0]
B0_3 = 0.33
B1_3 = -1.5

print('Curva 1:')
a1,b1,c1,d1 = cubic_spline_clamped_numeric(xs1, ys1, B0_1, B1_1)
spline_segmento(xs1, a1,b1,c1,d1, 'Curva 1')
plot_numeric_spline(xs1, ys1, a1,b1,c1,d1, 'Curva 1: Spline C√∫bico Clamped')

print('\nCurva 2:')
a2,b2,c2,d2 = cubic_spline_clamped_numeric(xs2, ys2, B0_2, B1_2)
spline_segmento(xs2, a2,b2,c2,d2, 'Curva 2')
plot_numeric_spline(xs2, ys2, a2,b2,c2,d2, 'Curva 2: Spline C√∫bico Clamped')

print('\nCurva 3:')
a3,b3,c3,d3 = cubic_spline_clamped_numeric(xs3, ys3, B0_3, B1_3)
spline_segmento(xs3, a3,b3,c3,d3, 'Curva 3')
plot_numeric_spline(xs3, ys3, a3,b3,c3,d3, 'Curva 3: Spline C√∫bico Clamped')
```
