**2. Encuentre soluciones precisas dentro de $10^{-4}$ para los siguientes problemas.**

```python
import numpy as np

def newton_raphson(f, df, p0, tol=1e-4, max_iter=50):
    pn = p0
    
    for i in range(1, max_iter + 1):
        try:
            df_pn = df(pn)
            if abs(df_pn) < 1e-10:
                print(f"Error: DivisiÃ³n por cero en la iteraciÃ³n {i}. f'(p{i-1}) = 0.")
                return None
                
            pn_next = pn - f(pn) / df_pn            
            error = abs(pn_next - pn)
            print(f"n={i:<2}, p_n={pn_next:.5f}, Error={error:.5e}")
            
            if error < tol:
                print(f"\n Convergencia alcanzada en {i} iteraciones.")
                return pn_next
            
            pn = pn_next
            
        except Exception as e:
            print(f"OcurriÃ³ un error: {e}")
            return None

    print(f"\n El mÃ©todo no convergiÃ³ despuÃ©s de {max_iter} iteraciones.")
    return pn
```

```python
import numpy as np
from scipy.optimize import newton

def secante(f, x0, x1, tol=1e-5, max_iter=50):
    iteracion = 0
    while abs(x1 - x0) > tol and iteracion < max_iter:
        fx0 = f(x0)
        fx1 = f(x1)
        if fx1 - fx0 == 0:
            print("Error: DivisiÃ³n por cero en la secante.")
            return None
        x2 = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
        x0, x1 = x1, x2
        iteracion += 1
    return x1
```

```python
import math

def f(x): return math.tan(math.pi * x) - 6
def df(x): return math.pi / math.cos(math.pi * x)**2

p_real = math.atan(6) / math.pi
print(f"RaÃ­z real â‰ˆ {p_real:.10f}\n")

def biseccion(a, b, n=10):
    print("MÃ©todo de BisecciÃ³n:")
    for i in range(1, n + 1):
        p = (a + b) / 2
        print(f"IteraciÃ³n {i}: a={a:.6f}, b={b:.6f}, p={p:.6f}")
        if f(a) * f(p) < 0:
            b = p
        else:
            a = p
    print(f"Resultado â‰ˆ {p:.10f}\n")
    return p

def newton(p0, n=10):
    print("MÃ©todo de Newton:")
    for i in range(1, n + 1):
        p = p0 - f(p0) / df(p0)
        print(f"IteraciÃ³n {i}: p={p:.6f}")
        p0 = p
    print(f"Resultado â‰ˆ {p:.10f}\n")
    return p

def secante(p0, p1, n=10):
    print("MÃ©todo de la Secante:")
    for i in range(1, n + 1):
        p = p1 - f(p1) * (p1 - p0) / (f(p1) - f(p0))
        print(f"IteraciÃ³n {i}: p={p:.6f}")
        p0, p1 = p1, p
    print(f"Resultado â‰ˆ {p:.10f}\n")
    return p
```

```python
import math

def f(x):
    return math.copysign(abs(x)**(1/3), x)   # raÃ­z cÃºbica real

def df(x):
    # derivada (1/3) x^{-2/3} (no definida en 0)
    if x == 0:
        return float('inf')
    return (1.0/3.0) * abs(x)**(-2.0/3.0)

# Newton empezando en p0 = 1
x = 1.0
print("Newton (p0 = 1):")
for k in range(10):
    d = df(x)
    if d == 0 or math.isinf(d):
        print("  derivada problemÃ¡tica; paro")
        break
    x = x - f(x)/d
    print(f"  IteraciÃ³n {k+1}: x = {x:.12f}")

# Secante con p0 = 5, p1 = 0.5
p0, p1 = 5.0, 0.5
print("\nSecante (p0 = 5, p1 = 0.5):")
for k in range(12):
    f0, f1 = f(p0), f(p1)
    denom = (f1 - f0)
    if abs(denom) < 1e-15:
        print("  Error. Denominador pequeÃ±o.")
        break
    p2 = p1 - f1*(p1 - p0)/denom
    print(f"  IteraciÃ³n {k+1}: p = {p2:.12f}")
    p0, p1 = p1, p2
```

Ambos mÃ©todos divergen. En el mÃ©todo de Newton los valores cambian mucho y el de la secante tambiÃ©n oscila, se dificulta la aproximaciÃ³n a la raÃ­z.
