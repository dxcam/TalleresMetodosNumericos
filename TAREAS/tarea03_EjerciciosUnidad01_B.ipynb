**1. Utilice aritm√©tica de corte de tres d√≠gitos para calcular las siguientes sumas. Para cada parte ¬øqu√© m√©todo es m√°s preciso y por qu√©?**


````python
import math

def cortar_tres_digitos(x):                     # Funci√≥n para truncar a 3 cifras significativas
    if x == 0:                                  # Si el n√∫mero es cero, retorna 0.0
        return 0.0
    exp = math.floor(math.log10(abs(x)))        # Calcula el orden de magnitud del n√∫mero
    factor = 10 ** (2 - exp)                    # Calcula el factor necesario para truncar el n√∫mero a 3 cifras significativas => 10^(3-1-exp)
    return math.trunc(x * factor) / factor      # Multiplica el n√∫mero por el factor, lo trunca (quita decimales sin redondear)

def suma_con_corte(serie, orden='asc'):                                 # Funci√≥n para sumar con truncamiento de 3 cifras significativas
    total = 0.0                                                         # Inicializa el total acumulado.
    datos = serie if orden == 'asc' else list(reversed(serie))          # Determina el orden de la serie: Si es descendente,la invierte.
    for x in datos:                                                     # Para cada elemento x, lo trunca, lo suma al total, y luego trunca el resultado.
        total = cortar_tres_digitos(total + cortar_tres_digitos(x))
    return total                                                        # Se devuelve el total truncado.

def mostrar_errores(valor_real, valor_aprox):
    error_abs = abs(valor_real - valor_aprox)
    error_rel = error_abs / abs(valor_real)
    print(f"Error absoluto: |{valor_real:.9f} - {valor_aprox:.9f}| = {error_abs:.9f}")
    print(f"Error relativo: |({valor_real:.9f} - {valor_aprox:.9f}) / {valor_real:.9f}| = {error_rel:.9f}\n")

sumatoria_a = [1 / (i**2) for i in range(1, 11)]
sumatoria_b = [1 / (i**3) for i in range(1, 11)]

a_asc = suma_con_corte(sumatoria_a, orden='asc')
a_desc = suma_con_corte(sumatoria_a, orden='desc')
b_asc = suma_con_corte(sumatoria_b, orden='asc')
b_desc = suma_con_corte(sumatoria_b, orden='desc')

suma_real_a = sum(1 / (i**2) for i in range(1, 11)) #Sumas sin corte
suma_real_b = sum(1 / (i**3) for i in range(1, 11))
````

````python
print(f"Sumatoria sin corte de 3 cifras significativas: {suma_real_a:.9f}")
print(f"Suma ascendente con corte: {a_asc}")
mostrar_errores(suma_real_a, a_asc)
print(f"Suma descendente con corte: {a_desc}")
mostrar_errores(suma_real_a, a_desc)
````

**b.** $\sum_{i=1}^{10} \frac{1}{i^3}$ primero por $\frac{1}{1} + \frac{1}{8} + \frac{1}{27}+ ‚ãØ + \frac{1}{1000}$ y luego por $\frac{1}{1000} + \frac{1}{729} + ‚ãØ + \frac{1}{1}$

````python
print(f"Sumatoria sin corte: {suma_real_b:.9f}")
print(f"Suma ascendente con corte: {b_asc}")
mostrar_errores(suma_real_b, b_asc)
print(f"Suma descendente con corte: {b_desc}")
mostrar_errores(suma_real_b, b_desc)
````

En ambos casos se puede observar que el segundo m√©todo posee un menor error absoluto, por lo tanto, es m√°s preciso. Al sumar de menor a mayor hay menos p√©rdida de cifras significativas, ya que estas cifras se conservan en las primeras sumas, brindando menor acumulaci√≥n del error de corte.

**5. a. ¬øCu√°ntas multiplicaciones y sumas se requieren para determinar una suma de la forma**
$$\sum_{i=1}^{n} \sum_{j=1}^{i} a_ib_j$$

$Multiplicaci√≥n= \sum_{i=1}^{n} i= \frac{n(n+1)}{2} $

$Suma = \sum_{i=1}^{n} (i-1)= \sum_{i=1}^{n} i-n= \frac{n(n+1)}{2} - n = \frac{n(n-1)}{2} $

````python
def contar_operacion(n):
    multiplicacion = n * (n + 1) // 2
    suma = n * (n - 1) // 2
    return multiplicacion, suma

# Ejemplo de uso
n = 8
multiplicacion, suma = contar_operacion(n)
print(f"Para n = {n}, se necesitan {multiplicacion} multiplicaciones y {suma} sumas.")
````

**b. Modifique la suma en la parte a) a un formato equivalente que reduzca el n√∫mero de c√°lculos.**


````python
def suma_optimizada(a, b):
    n = len(a)
    suma_acumulada = [0] * n
    suma_acumulada[-1] = a[-1]

    for i in range(n - 2, -1, -1):                              # Precomputar sumas acumuladas desde el final
        suma_acumulada[i] = suma_acumulada[i + 1] + a[i]

    resultado = 0
    for j in range(n):                              # Calcular la suma optimizada
        resultado += b[j] * suma_acumulada[j]

    return resultado

# Ejemplo
a = [2, 4, 6, 8, 10]
b = [1, 3, 5, 7, 9]
resultado = suma_optimizada(a, b)
print("Resultado de la suma optimizada:", resultado)
````

**3. Suponga que**
$$\frac{1‚àí2ùë•}{1‚àíùë•+ùë•^2} + \frac{2ùë•‚àí4ùë•^3}{1‚àíùë•^2+ùë•^4} + \frac{4ùë•^3 ‚àí 8ùë•^7}{1 ‚àí ùë•^4 + ùë•^8} + ‚ãØ = \frac{1+2ùë•}{1+ùë•+ùë•^2} ,$$
**para ùë• < 1 y si ùë• = 0.25. Escriba y ejecute un algoritmo que determine el n√∫mero de t√©rminos necesarios en el lado izquierdo de la ecuaci√≥n de tal forma que el lado izquierdo difiera del lado derecho en menos de $10^{‚àí6}$**



```
ENTRADA: x = 0.25.
SALIDA: k 

Paso 1: Determine tolerancia = 1e-6.
Paso 2: Calcule lado_derecho = (1 + 2x) / (1 + x + x¬≤).
Paso 3: Determine suma_lado_izquierdo = 0.
Paso 4: Determine k = 0.
Paso 5: Repita
            Determine numerador = 2^k * x^(2^k ‚àí 1) ‚àí 2^(k+1) * x^(2^(k+1) ‚àí 1).
            Determine denominador = 1 ‚àí x^(2^k) + x^(2^(k+1)).
            Determine t√©rmino = numerador / denominador.
            Determine suma_lado_izquierdo = suma_lado_izquierdo + t√©rmino.
            Determine k = k + 1.
        Hasta que |suma_lado_izquierdo ‚àí lado_derecho| < tolerancia.
Paso 6: SALIDA k;
        PARE
FIN
```

````python
x = 0.25
tolerancia = 1e-6
lado_derecho = (1 + 2 * x) / (1 + x + x**2) # Calcular el lado derecho de la ecuaci√≥n

suma_lado_izquierdo = 0
k = 0

while abs(suma_lado_izquierdo - lado_derecho) >= tolerancia:                    # Bucle para sumar t√©rminos hasta alcanzar la tolerancia
    numerador = (2**k) * (x**(2**k - 1)) - (2**(k + 1)) * (x**(2**(k + 1) - 1))
    denominador = 1 - x**(2**k) + x**(2**(k + 1))
    termino = numerador / denominador
    suma_lado_izquierdo += termino
    k += 1

print("N√∫mero de t√©rminos necesarios en el lado izquierdo:", k)
````
